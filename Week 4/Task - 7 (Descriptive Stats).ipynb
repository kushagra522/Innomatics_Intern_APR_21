# Mean


**The mean (arithmetic mean) is the average (sum of observations / number of observations) in a distribution, sample or population.**

**CONVENTIONAL SYMBOLS**: <br>

1) POPULATION MEAN : 


## $\mu$ = $\frac{\sum x}{N}$

2) SAMPLE MEAN :

## $\bar{x}$  = $\frac{\sum x}{n}$


**WHERE** <br>
$\sum x~$ : Sum of values of all observation<br>
N $~~~~$       : Number of elements in the population<br>
n $~~~~~$       : Number of elements in the sample <br>


NOTE : <br>
1) If our data contains outliers mean will get affected.

```python
## MEAN USING CODING 
"""Let us find the mean of Annual_HH_Income from data"""
total = data['Annual_HH_Income'].sum()
n = len(data['Annual_HH_Income'])
print("Mean of the data is :",total/n)

#USING PANDAS LIBRARY:
print("Mean of the data using pandas function is", data['Annual_HH_Income'].mean())
```

# Median
**Median is single value from data set that measure the central item in data, it gives the _middlemost_ or _most central item_ in the given data.**

1) Arrange the data in ascending or descending order.
2) If the data contains odd number of items , the middle element of the data is median
3) If the data contains even number then median is average of two middle items.

In formal language , the median is : $\frac{n+1}{2}$th item in a data.

NOTE : <br>
1) Outliers doesn't affect mean, hence we can use this in a data having outliers.<br>
2) The disadvantage is that median is more complex than mean, its time complexity is more.

```python
#MEDIAN USING CODE
"""Let's find the median of Annual Household income from data"""

#STEP 1: Arrange in ascending/descending order
array = data['Annual_HH_Income'].sort_values().reset_index(drop=True)

#STEP 2 : Finding the nth element
"""Since our N= 50 which is even."""
#Hence middle element will be mean of 24th and 25th element

print("Median of data is :", (array[24]+array[25])/2)


#USing Pandas function
print("Median of data using pandas function is",data['Annual_HH_Income'].median())
```

## Dispersion : Why its important?



**As we can see the mean of all three curves is the same, but curve A has less spread
(or variability) than curve B, and curve B has less variability than curve C. If we measure only the mean
of these three distributions, we will miss an important difference among the three curves.**

To increase our understanding of the pattern of the data, we must also measure its
dispersionits spread, or variability.

1) It gives us additional information that enables us to judge the reliability of our measure of the central tendency. If data are widely dispersed,such as those in curve C in ,the central location is less representative of the data as a whole than it would be for data more closely centered around the mean, as in curve A. 

2) Because there are problems peculiar to widely dispersed data, we must be able to recognize that data are widely dispersed before we can tackle those problems. 

3) We may wish to compare dispersions of various samples. If a wide spread of values away from the center is undesirable or presents an unacceptable risk, we need to be able to recognize and avoid choosing the distributions with the greatest dispersion.

## Variance



**The measure of deviation from the central tendency is called variance.**

**Population Variance**


WHERE:<br>
$\sigma^{2}$ = population variance<br>
x = item or observation<br>
$\mu$ = population mean<br>
N = total number of items in population<br>
$\sum$ = sum of all the values (x  $\mu)^{2}$.





**Sample variance**<br>

WHERE:<br>
$s^{2}$ = Population variance<br>
x = Item or observation<br>
$\bar{x}$ = Mean of sample<br>
n = Total number of items in sample<br>
  = Sum of all the values $(x   \bar{x})^{2}$ .<br>

NOTE: <br>
1) For variance, the units are the sqaures of the units of the data, hence we use squared root which is called standard deviation.

# Standard Deviation

**Since we cannot use the sqaure of unit we take the sqaure root variance, which is called standard deviation.**<br>

Population standard deviation:<br>
$\sigma = \sqrt[]{\sigma^{2}}$

Sample standard deviation:<br>
s = $\sqrt[]{s^{2}}$

```python
# VARIANCE AND STANDARD DEVIATION USING CODING
"""Lets calculate the variance and standard deviation for the column Monthly House-Hold Income"""
#Lets first visualize the distribution of data
import seaborn as sns
sns.distplot(data['Mthly_HH_Income'])
```

As we can see the spread of data is uneven, and there lies two peaks thus its important to calculate the variance for this data.

```python
#Using variance formula 
#Step 1 : To calculated mean
mean = data['Mthly_HH_Income'].sum()/len(data['Mthly_HH_Income'])
#Step 2 : Difference between observed and mean value:
diff_obs_and_mean = data['Mthly_HH_Income'] - mean
#Step 3 : Squaring the diff_m values
square_of_diff_obs_and_mean = diff_obs_and_mean**2
#Step 4 : Dividing the sum by total number of values called variance.
N = 50
variance = square_of_diff_obs_and_mean.sum()/49
print("Variance of the given data is :", variance)

print(80*'-')
#Using pandas built in variance
print("Variance of the given data using pandas built in function is", data['Mthly_HH_Income'].var())
print(80*'-')
print("Note: Since the given data is sample and not population data, hence its the sample variance.")
```

```python
#Standard deviation 
print("The sample standard deviation is", variance**0.5)
print(80*'-')
print("The sample standard deviation using pandas built in function is ",data['Mthly_HH_Income'].std())
```

# Positively Skewed & Negatively Skewed Normal Distribution

Skewed Distributions : They are called skewed because values in their frequency distributions are concentrated
at either the low end or the high end of the measuring scale on the horizontal axis. The values
are not equally distributed.As in the following image.


1) Curve A is skewed to the right (or positively skewed) because it tails off toward the high end of the scale.

2) Curve B is skewed to the left (negatively skewed) because it tails off toward the low end of the scale.

**How skewness affects mean , median or mode?**<br>
When the population is skewed negatively or positively, the median is often the best measure of location because it is always between the mean and the mode.

The median is not as highly influenced by the frequency of occurrence of a single value as is the mode, nor is it pulled by extreme values as is the mean.

# QQ-Plot

To understand QQ-Plot we need to understand first the term ***Quantile***.<br>
The word ***quantile*** comes from the word quantity. In simple terms, a quantile is where a sample is divided into equal-sized, adjacent, subgroups.It can also refer to dividing a probability distribution into areas of equal probability.<br>
We mainly use quantile types:

1) **Quartile** which divides the group into four equal parts.

2) **Percentile** which divides the group into hundred equal parts.

3) **Decile** which divides the group into ten equal parts.

Step to find quartules:<br>
Step 1: Order the data from smallest to largest.<br>
Step 2: Count how many observations you have in your data set.<br>
Step 3: Convert any percentage to a decimal for q. Eg.If we are looking for the number where 20 percent of the values fall below it, so convert that to 0.2.<br>
Step 4: Insert your values into the formula: ith observation = q (n + 1)

```python
#Lets find the quartile value for monthly household expenses.

hh_expense_arranged = data['Mthly_HH_Expense'].sort_values().reset_index(drop=True) #Sorting

total_obs = len(data['Mthly_HH_Expense']) # Total observation

q = 0.25 #finding 25 percentile

index = int(0.25*(total_obs+1)) #Finding the nth term

print("The 25th percentile value for data is: ",hh_expense_arranged[index])
```

```python
#Using pandas builtin function
print("The 25th percentile value for data is using pandas function is",data['Mthly_HH_Expense'].quantile(q=0.25))
```

**Why do we need to use quantiles?**<br>
Quantiles give some information about the shape of a distribution - in particular whether a distribution is skewed or not.

***QQ PLOT*** is a plot of quantiles of two distributions against each other, or a plot based on estimates of quantiles.
A pattern of plots is used to compare the two distribution. A 45 reference line is also plotted.
If the distribution of data falls approx on the line , then we can say that the given distribution is normal.

```python
uniform_100 = np.random.uniform(low = 10, high = 40, size = (100))
normal_100 = np.random.normal(loc = 20, scale = 5, size=100)
df_100 = pd.DataFrame({'uniform' : uniform_100,
                  'normal' : normal_100})
stats.probplot(df_100['normal'], dist="norm", plot=plt)
plt.grid()
```

This is how QQ-plot for normal distribution is proved

```python
# Lets try to check normality of the column Annual Household income column from our data.
stats.probplot(data['Annual_HH_Income'], dist="norm", plot=plt)
plt.grid()
```

**As we can see that distribution doesn't follow on the line , hence we can conclude that the Annual household income don't follow normal distribution.**

# Box-Cox Plot

**Box-Cox** tranformation plot is generally applied to convert ***Power Law*** or ***Pareto*** curves , to transform into the normal distribution.

It plots the correlation coeffient for various coefficient for various values of lambda.

```python
#Lets try to look how pareto curve looks
pareto_rv = np.random.pareto(a = 2, size = 1000)
sns.distplot(pareto_rv)
```

```python
# Lets check its normality using QQ-Plot
stats.probplot(pareto_rv, dist="norm", plot=plt)
plt.grid()
```

Its clearly not a normal distribution


```python
#Lets tranform it using box-cox and check the normality.
data_trans, l = stats.boxcox(pareto_rv)
stats.probplot(data_trans, dist="norm", plot=plt)
plt.grid()
```

Our data has successfully converted to normal distribution
