```python
jobs = ['programmer analyst', 'software engineer ', 'hardware engineer','associate software engineer' , 'associate system engineer'  ,'associate engineer']
d = data['Salary'].loc[(data['Designation']=='programmer analyst')|(data['Designation']=='software engineer')|(data['Designation']=='hardware engineer')|(data['Designation']=='associate engineer')]
```

```python
f, axes = plt.subplots(1,2)
f.set_figheight(5)
f.set_figwidth(15)
sns.boxplot(d,ax=axes[0])
sns.distplot(d,ax=axes[1])
```

```python
d = d.reset_index(drop=True)
```

```python
Q3 = d.quantile(0.75)
print("Q3 quartile = ", Q3)
Q1 = d.quantile(0.25)
print("Q1 quartile =", Q1)
IQR = Q3 - Q1
print("IQR =", IQR)
print("Data above Q3 =",(Q3) + 1.5*IQR)
print("Data below Q1 =",(Q1) - 1.5*IQR)
```

```python
print(len(d[d>550000]))
print(len(d[d>750000]))
```

```python
d_refined = d[d<=550000]
print(d_refined.shape)
```

```python
print('Mode of the data',d_refined.mode())
print('Mean of the data',d_refined.mean())
print('Median of the data',d_refined.median())

```

Alternate Hypothesis (Bold claim): $$ H_1:~~avg~package<3,00,000 $$
Null Hypothesis (Status Quo)     : $$ H_0:~~avg~package>=3,00,000 $$ 

Using <br>
$$ z = \frac{\bar{x} - \mu}{\sigma/\sqrt[2]{n}} $$

```python
from scipy.stats import norm
import statistics as st
data = d_refined

n = d.shape[0] #No of observation

x_bar = 300000 #Given mean value

mu = st.mean(data) #sample mean
print("Mean of data is :", mu)

std = st.stdev(data) #sample std deviation
print("Std.deviation of data is :",std )

z_score = (x_bar-mu)/(std/((n)**0.5)) #tscore value
print("z-score value is :",z_score)

confidence_level = 0.95
alpha = 1- confidence_level

z_critical = norm.ppf(1-alpha)
print("z_critical value is :",z_critical)
print(40*'*')
if z_critical>z_score:
    print("Failed to reject the null hypothesis")
else:
    print("Reject Null Hypothesis")
print(40*'*')    
```

***Conclusion : With 95% confidence , we can say that we failed to reject the null hypothesis and the average package for the given designations is upto 3 lakhs.***

```python
#Is there a relationship between gender and specialisation?
from scipy.stats import chi2
from scipy.stats import chi2_contingency

data['Gender'].value_counts()
```

```python
data['Specialization'].value_counts()
```

Lets make a bold Claim that Gender and Specialization are dependent.

Alternate Hypothesis (Bold claim): $$ H_1:~~They~are~dependent $$
Null Hypothesis (Status Quo)     : $$ H_0:~~They~are~not~dependent $$ 

```python
#Looking at the freqency distribution

observed = pd.crosstab(data.Specialization, data.Gender)
observed.head()
```

```python
# chi2_contigency returns chi2 test statistic, p-value, degree of freedoms, expected frequencies
# Computing chi2 test statistic, p-value, degree of freedoms

chi2_test_stat = chi2_contingency(observed)[0]
pval = chi2_contingency(observed)[1]
df = chi2_contingency(observed)[2]
print("The chi sqaure score is :" , chi2_test_stat)
```

```python
confidence_level = 0.95

alpha = 1 - confidence_level

chi2_critical = chi2.ppf(1 - alpha, df)

chi2_critical
```

```python
if(chi2_test_stat > chi2_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

***Conclusion : With 95% confidence interval we failed to reject null hypothesis and we can say that the Gender and Specialization are dependent.***
